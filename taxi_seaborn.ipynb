```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
taxi_data = sns.load_dataset('taxis')

taxi_data.to_csv('taxi_dataset.csv', index=False)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=taxi_data['total'])
plt.title('Boxplot der Spalte "total"')
plt.show()
```
